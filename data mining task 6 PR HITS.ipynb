```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G1= nx.erdos_renyi_graph(20,0.2,seed=42,directed=True) 
nx.draw(G1, with_labels=True) 
plt.show() 
```

```python
pr=nx.pagerank(G1,0.4)
pr
```

```python
G2=nx.barabasi_albert_graph(20,11,seed=41) 
nx.draw(G2, with_labels=True) 
plt.show()
```

```python
pr=nx.pagerank(G2,0.4) 
pr 
```

```python
hubs, authorities = nx.hits(G1, max_iter = 50, normalized = True) 
  
print("Hub Scores: ", hubs) 
print("Authority Scores: ", authorities)
```

```python
hubs, authorities = nx.hits(G2, max_iter = 100, normalized = True) 
# The in-built hits function returns two dictionaries keyed by nodes 
# containing hub scores and authority scores respectively. 
  
print("Hub Scores: ", hubs) 
print("Authority Scores: ", authorities)
```
